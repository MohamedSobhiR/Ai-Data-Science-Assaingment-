## getting current tempreture

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.timeanddate.com/weather/egypt/cairo'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
temp = soup.find('div', attrs={'class': 'h2'}).get_text()
temp
```

```python
current_temp = temp.split( "\xa0" )[0]
int(current_temp)
```

```python
sendy['Temperature'].fillna(sendy['Temperature'].median() , inplace = True ) # Fill Nan Values with median 
```

```python
sendy.info()
```

```python
sendy['Temp_Diff'] = sendy['Temperature'].apply(lambda x: x -int(current_temp))
```

```python
sendy['Temp_Diff']
```

```python
sendy.info()
```

```python
def c_To_f (x):
    return x * (9/5) + 32  #20Â°C Ã— 9/5 + 32 = 68 Â°F
```

```python
sendy['Fehr_Temp'] = sendy['Temperature'].apply(c_To_f)
```
