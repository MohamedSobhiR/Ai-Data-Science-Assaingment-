```python
df['Age'].unique()
```

```python
age = {'0-17':'1' , '18-25':'2', '26-35':'3', '36-45':'4', '46-50':'5', '51-55':'6', '55+':'7'}
#using map  to  modify age columns 
df['Age'] = df['Age'].map(age)
df['Age']
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```

```python
df['Stay_In_Current_City_Years'].value_counts()
```

```python
df['Stay_In_Current_City_Years'] = pd.to_numeric(df['Stay_In_Current_City_Years'], errors='coerce')
```

```python
df['Stay_In_Current_City_Years'].fillna(4, inplace=True)
```

```python
df.info()
```

```python
df['Product_Category_2'].fillna(0, inplace=True)
df['Product_Category_3'].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df
```

```python
city = {'A':'1' , 'B':'2', 'C':'3'}
gender = {'F':'0' , 'M':'1'}
#using map  to  modify age columns 
df['Gender'] = df['Gender'].map(gender)
df['City_Category'] = df['City_Category'].map(city)
```

```python
df.drop('User_ID', axis=1, inplace=True)
df.drop('Product_ID', axis=1, inplace=True)
df
```

```python
df.describe()
```

```python
x = df.drop('Purchase', axis=1)
x
```

```python
y = df['Purchase']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
scaler= StandardScaler()

scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
models = {
    "LR": LinearRegression(),
    "KNNR" : KNeighborsRegressor(), 
    "DT": DecisionTreeRegressor(),
    "RF": RandomForestRegressor(),
    "XGBR": XGBRegressor(),
    "SVR": SVR()
}
```

```python
for name, model in models.items():
    print(f'Using model: {name}')
    model.fit(x_train, y_train)
    print(f'Training Score: {model.score(x_train, y_train)}')
    print(f'Test Score: {model.score(x_test, y_test)}')  
    print('-'*10)
```

```python
model = XGBRegressor()

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```
