````python
import sqlite3
````

````python
connection = sqlite3.connect('BookTOscrapDataBase.db')
cursor = connection.cursor()
````

````python
def get_rows(sql, data):
    global connection
    global cursor
    cursor.execute(sql, data)
    rows = cursor.fetchall()
    for BookID, BookName, BookPrice, BookRate, CategoryID in rows:
        print(f"[{BookID}] -_-_- {BookName} -_-_- {BookPrice} £-_-_- Rate :  {BookRate} ----- Cat_id: {CategoryID}")
````

```01) Write a sql query to get books that has rate >=3 and has 'Mr' in its name.```

````python
get_rows("SELECT * FROM Books WHERE BookRate >= ? AND BookName LIKE ?", [3, '%Mr%'])
````

```02) Write a sql query to get books that has rate >=3 or price > 20.```

````python
get_rows("SELECT * FROM Books WHERE BookRate >= ? OR BookPrice > ? ", [3,20])
````

**15) Write a sql query to calculate how many books each rate has and have price between 20£ and 30£.**


````python
cursor.execute("SELECT  COUNT(BookPrice) , BookRate  FROM Books  WHERE BookPrice BETWEEN ? and ? GROUP BY BookRate ", [20,30])
rows = cursor.fetchall()
print (rows)
````

**16) Write a sql query to calculate how many books each category has having count > 10.**


````python
cursor.execute("SELECT  COUNT(BookID) , CategoryID  FROM Books  GROUP BY CategoryID HAVING COUNT(BookID) > ? ", [10])
rows = cursor.fetchall()
print (rows)
````
