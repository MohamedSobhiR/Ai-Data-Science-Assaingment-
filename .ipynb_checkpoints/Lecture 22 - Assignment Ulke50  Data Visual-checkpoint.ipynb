```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as mp
```

```python
%matplotlib inline
sns.set(rc={'figure.figsize': [12, 12]}, font_scale=1.5)
```

```python
ulke = pd.read_csv('50ulke.csv')
ulke.head()
```

```python
ulke.info()
```

```python
ulke.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
```

```python
ulke.describe()
```

```python
sns.heatmap(ulke.corr(), annot=True)
```

```python
sns.distplot(ulke['Per person'] , kde = False , color = 'm')
```

```python
sns.distplot(ulke['National Income'] , kde = False , color = 'm')
```

```python
sns.distplot(ulke['Growth'] ,  color = 'm')
```

```python
sns.countplot(x='Continent' , data = ulke ,palette='rocket')
```

```python
sns.barplot(x='Continent', y='National Income', data=ulke , )
```

```python
sns.barplot(x='Continent', y='Per person', data=ulke , )
```

```python
sns.barplot(x='Continent', y='Growth', data=ulke , )
```

```python
sns.jointplot(x = 'Growth', y = 'National Income' , data = ulke   )
```

```python
sns.jointplot(x = 'Growth', y = 'Per person' , data = ulke, kind='kde', color='m')
```

```python
sns.boxplot(x="Continent", y="Per person", data=ulke)
```

```python
sns.boxplot(x="Continent", y="Growth", data=ulke)
```

```python
sns.lmplot(x="Per person", y="Growth", data=ulke, height=8, aspect=1.5)
```

```python
sns.lmplot( x='Per person', y='National Income',data=ulke, height=8, aspect=1.5)
```

```python
sns.lmplot(x='Growth',y='National Income',  data=ulke, height=8, aspect=1.5)
```
