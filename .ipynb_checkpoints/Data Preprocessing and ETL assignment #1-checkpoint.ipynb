## getting current tempreture

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.timeanddate.com/weather/egypt/cairo'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
temp = soup.find('div', attrs={'class': 'h2'}).get_text()
temp
```

```python
current_temp = temp.split( "\xa0" )[0]
int(current_temp)
```

```python
sendy['Temperature'].fillna(sendy['Temperature'].median() , inplace = True ) # Fill Nan Values with median 
```

```python
sendy.info()
```

```python
sendy['Temp_Diff'] = sendy['Temperature'].apply(lambda x: x -int(current_temp))
```

```python
sendy['Temp_Diff']
```

```python
sendy.info()
```

```python
def c_To_f (x):
    return x * (9/5) + 32  #20°C × 9/5 + 32 = 68 °F
```

```python
sendy['Fehr_Temp'] = sendy['Temperature'].apply(c_To_f)
```

```python
sendy['Fehr_Temp'] 
```

1) Using ufo.csv dataset, extract Month from 'Time' and use it to create new column and call it 'IN_December', then it will have a value of '1' if the month is December and '0' otherwise.


2) Using ufo.csv dataset, create new column and call it 'Time_Diff_From_WW1' that have time difference in years from the 'Time' Column and World War 1 start date that is 'July 28, 1914'.

```python
df = pd.read_csv('datasets/ufo.csv')
df.head()
```

```python
df.info()
```

```python
df['Time'] =  pd.to_datetime(df['Time'] , format = '%m/%d/%Y %H:%M' , errors = "coerce")
df.info()
```

```python
df.head()
```

```python
df['Month_Name'] = df['Time'].dt.month_name()
df.head()
```

```python
df['Month_Name'].value_counts()
```

```python
def in_december(x):
    if x == 'December' : 
        return '1'
    else :
        return '0'
```

```python
df['IN_December'] = df['Month_Name'].apply(in_december)
df.head()
```

## 2) 
Using ufo.csv dataset, create new column and call it 'Time_Diff_From_WW1' that have time difference in years from the 'Time' Column and World War 1 start date that is 'July 28, 1914'.

```python
from datetime import datetime
```

```python
time_of_ww1 = datetime( 1914 ,7, 28)
```

```python
df['Time_Diff_From_WW1'] =  (df['Time'] - time_of_ww1) / np.timedelta64(1, 'Y')
```

## Task 3
1) Using titanic_full.csv dataset, extract Mr,Miss,Mrs,etc.. from 'Name' column and use it to create new column and call it 'Title', then plot a countplot

showing the count for each title in the dataset.

2) Using sendy_logistics.csv extract user_id in a seperate column.

```python
import pandas as pd
import numpy as np 
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize':(8 , 8)}, font_scale=1.3)
```

```python
df = pd.read_csv('datasets/titanic_full.csv')
```

```python
df.head()
```

```python
def name_title (x):
    if 'Mrs' in x :
        return 'Mrs'
    elif 'Miss'in x :
        return 'Miss'
    elif 'Mr.'in x :
        return 'Mr'
    else :
        return 'Untitled'
```

```python
df['Title'] = df['Name'].apply(name_title)
```

```python
df
```

```python
sns.countplot(x= 'Title' , data = df)
```
